```python
df_clean['treatment'] = df_clean['treatment'].map({'Yes': 1, 'No': 0})
```

```python
df_clean['treatment'].value_counts()
```

```python
## df_clean['Age'] > 15. That means any age strictly greater than 15 is kept, so the smallest age that remains is 16. It’s just a way to remove any ages 15 or below, which are likely invalid for your dataset context (e.g., working-age adults).##
                                                                        
```

```python
 df_clean = df_clean[(df_clean['Age'] > 15) & (df_clean['Age'] < 100)]
```

```python
df_clean['Age'].describe()
```

```python
df_clean.head()
```

```python
df_clean.columns
```

```python
age_treated = df_clean[df_clean['treatment']==1]['Age']
age_not_treated = df_clean[df_clean['treatment']==0]['Age']
```

```python
from scipy.stats import shapiro

print("Shapiro test for treated group:", shapiro(age_treated))
print("Shapiro test for not treated group:", shapiro(age_not_treated))
```

```python
from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(age_treated, age_not_treated, equal_var=False)  # Welch’s t-test
print("T-statistic:", t_stat)
print("p-value:", p_val)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.boxplot(x='treatment', y='Age', data=df_clean)
sns.stripplot(x='treatment', y='Age', data=df_clean, color='black', alpha=0.3, jitter=True)
plt.xticks([0,1], ['Not Treated', 'Treated'])
plt.title('Age distribution by Treatment')
plt.ylabel('Age')
plt.xlabel('Treatment')
plt.show()
```

```python
##T-statistic = 2.61 → the difference between the mean ages of treated vs not-treated is 2.61 standard errors apart. p-value = 0.0091 → this is less than 0.05, so we reject the null hypothesis. Conclusion: There is a statistically significant difference in age between people who seek treatment and those who do not. Age does appear to affect the likelihood of treatment in this dataset.##
```
